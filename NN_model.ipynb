import necessary libraries and import the data set

```python
import pandas as pd

data = pd.read_csv('model_data.csv')
```

```python
print(data.columns)

data.drop('Unnamed: 0', axis=1, inplace=True)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pearson correlation by default in pandas
corr = data.corr()

# Seaborn Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, vmin=-1, vmax=1, center=0, annot=True, cmap= 'RdBu', fmt='.2f')
```

```python
# generating pairplot
sns.pairplot(data)
```
